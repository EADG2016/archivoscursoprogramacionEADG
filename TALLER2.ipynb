Dando continuidad con la primera entrega del proyecto, en esta oportunidad el estudiante debe realizar las siguientes validaciones utilizando la sentencia condicional IF.

Si el empleado es mayor de 55 años disfrutará de un bono de prepensión correspondiente al 5% de su sueldo básico.

Si el empleado es casado y tiene hijos se le otorgará un paseo cada diciembre

Si el sueldo básico está entre 1000000 y 1500000 tendrá una comisión del 2% sobre el valor del sueldo; Si el sueldo básico está entre 1500001 y 2000000 tendrá una comisión del 5% sobre el valor del sueldo; para todos los demás casos no habrá comisión.

Si el empleado trabajó más de 20 días al mes y su sueldo es menor a 1000000 tendrá derecho a un bono de alimentación.

```python
N_de_Identificación = input("Numero de identificación: ")
Nombres = input("Nombres: ")
Apellidos = input("Apellidos: ")
Dirección = input("Dirección:")
telefono = input("Telefono: ")
Edad = int(input("Edad: "))
Estado_civil = input ("Estado civil: ")
Numero_de_hijos = int(input("Numero de hijos: "))
Estatura_en_centímetros = input("Estatura en cm: ")
fecha_de_contratación = input("Fecha de contratación dia/mes/año: ")
Sueldo_basico = float(input("Sueldo basico: "))
Días_laborados = int(input("Días laborados al mes: "))

#Si el empleado es mayor de 55 años disfrutará de un bono de prepensión correspondiente al 5% de su sueldo básico.
if Edad > 55:
  bono = Sueldo_basico * 0.05
  Sueldo_total = Sueldo_basico + bono
  print("\nBono prepensión ${:,.0f} \nSueldo_total ${:,.0f}".format(bono, Sueldo_total))
else:
  print(f"\nSueldo total ${Sueldo_basico:,.0f}")

#Si el empleado es casado y tiene hijos se le otorgará un paseo cada diciembre
if Estado_civil == "Casado" and Numero_de_hijos >= 1:
  print("\nAcreedor de un viaje todos los Diciembres")
else:
  print("\nNo es aceedor de un viaje los Diciembres")

#Si el sueldo básico está entre 1000000 y 1500000 tendrá una comisión del 2% sobre el valor del sueldo
#Si el sueldo básico está entre 1500001 y 2000000 tendrá una comisión del 5% sobre el valor del sueldo; para todos los demás casos no habrá comisión.
if Sueldo_basico >= 1000000 and Sueldo_basico <=1500000:
  comi1 = Sueldo_basico * 0.02
  sueldo_comi1 = Sueldo_basico + comi1
  print("\nValor comisión ${:,.0f} \nSueldo Total mas comisión ${:,.0f}".format(comi1, sueldo_comi1))
if Sueldo_basico >= 1500001 and Sueldo_basico <=2000000:
  comi2 = Sueldo_basico * 0.05
  sueldo_comi2 = Sueldo_basico + comi2
  print("\nValor comisión ${:,.0f} \nSueldo Total mas comisión ${:,.0f}".format(comi2, sueldo_comi2))
else:
  print("No recibe comisión")

#Si el empleado trabajó más de 20 días al mes y su sueldo es menor a 1000000 tendrá derecho a un bono de alimentación
if Días_laborados > 20 and Sueldo_basico < 1000000:
  print("\nRecibe beneficio de bono de alimentación")
else:
  print("\nNo aplica para beneficio de bono de alimentación")
```
