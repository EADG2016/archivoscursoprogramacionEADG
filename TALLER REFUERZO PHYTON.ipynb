1. Un vendedor recibe un sueldo base más un 10% extra por
comisión de sus ventas, el vendedor desea saber cuánto
dinero obtendrá por concepto de comisiones por las tres
ventas que realiza en el mes y el total que recibirá en
el mes tomando en cuenta su sueldo base y comisiones.

```python
sueldo_base = float(input("Digite el sueldo base: $"))
venta1 = float(input("Digite el valor de la venta 1: $"))
venta2 = float(input("Digite el valor de la venta 2: $"))
venta3 = float(input("Digite el valor de la venta 3: $"))
porc_extra = (venta1 + venta2 + venta3) * 0.1
Total_sueldo = (porc_extra + sueldo_base)

print("\nEl sueldo base es = ${:,.0f}".format(sueldo_base))
print("\nLas comisiones son = ${:,.0f}".format(porc_extra))
print("\nEl sueldo total es = ${:,.0f}".format(Total_sueldo))
```

2. Una tienda ofrece un descuento del 15% sobre el total
de la compra y un cliente desea saber cuánto deberá
pagar finalmente por su compra.

```python
total_compra = float(input("Digite el valor de la compra sin descuento: $"))

porc_descuento = round(total_compra * 0.15)
valor_a_pagar = (total_compra - porc_descuento)
print("\nEl descuento en tu compra es de = ${:,.0f}".format(porc_descuento))
print("\nEl valor a pagar de tu compra es de = ${:,.0f}".format(valor_a_pagar))
```

3. Un alumno desea saber cuál será su calificación final en
la materia de Algoritmos. Dicha calificación se compone
de tres exámenes parciales.

```python
cal1 = float(input("Digite la calificaicón 1: "))
cal2 = float(input("Digite la calificaicón 2: "))
cal3 = float(input("Digite la calificaicón 3: "))

prom = round((cal1 + cal2 + cal3) / 3, 2)

print(f"El promedio es= {prom}")
```

4. Un maestro desea saber qué porcentaje de hombres y que
porcentaje de mujeres hay en un grupo de estudiantes.

```python
num_hombres = float(input("Digite número de hombres: "))
num_mujeres = float(input("Digite número de mujeres: "))

porchom = round((num_hombres) / (num_hombres + num_mujeres) * 100, 2)
print(f"El porcentaje de hombres es= {porchom} %")
porchom = round((num_mujeres) / (num_hombres + num_mujeres) * 100, 2)
print(f"El porcentaje de mujetes es= {porchom} %")
```
