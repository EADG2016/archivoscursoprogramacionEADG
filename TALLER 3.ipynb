```python
def info (id, name, lastname, addr, phone, age, ec, nson, est, fecha_contra, sueldo, diaslab):
  print(f"Numero de identificación: {id}")
  print(f"Nombres: {name}")
  print(f"Apellidos: {lastname}")
  print(f"Dirección: {addr}")
  print(f"Telefono: {phone}")
  print(f"Edad: {age}")
  print(f"Estado civil: {ec}")
  print(f"Numero de hijos: {nson}")
  print(f"Estatura en cm: {est}")
  print(f"Fecha de contratación (dia/mes/año): {fecha_contra}")
  print(f"Sueldo: {sueldo}")
  print(f"Dias laborados al mes: {diaslab}")

resp = "Si"

while resp =="si" or resp == "Si" or resp == "SI":
  ident = input("Digite su numero de identificación: ")
  nombres = input("Digite sus Nombres: ")
  apellidos = input("Digite sus apellidos: ")
  direccion = input("Digite su dirección: ")
  telefono = input("Digite su telefono: ")
  edad = input("Digite su Edad: ")
  Estado_civ = input("Cual es su estado civil: ")
  Num_hijos = input("Digite su número de hijos: ")
  Estatura = input("Digite su estatura en cm: ")
  Fecha_con = input("Digite su fecha de contratación (dia/mes/año): ")
  Sueldo = input("Digite su sueldo básico: $")
  Dias_lab = input("Digite los dias laborados al mes")
  info(ident, nombres, apellidos, direccion, telefono, edad, Estado_civ, Num_hijos, Estatura, Fecha_con, Sueldo, Dias_lab)
  resp = input("Desea ingresar otro trabajador")
```
